```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("C:\\Users\\User\\Desktop\\Datasets\\Wine.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# distingguishing red and white wine

w = data["Flavanoids"].values
r = data["Proanthocyanins"].values

result = []
for i in range(len(w)):
    if w[i] > r[i]:
        result.append(0) # high flavanoids result in white colour
    else:
        result.append(1) # high proanthocyanins results in red colour

data["Colour"] = result
data["Colour"].values # 0 means white wine. 1 means red wine
```

```python
data
```

```python
X = data.drop(columns = ["Colour"])
Y = data["Colour"]
```

```python
# plotting original data
# Scatter plot before applying PCA
plt.figure(figsize=(10, 6))
plt.scatter(X[Y == 0]["Flavanoids"], X[Y == 0]["Proanthocyanins"], label = "Colour 0")
plt.scatter(X[Y == 1]["Flavanoids"], X[Y == 1]["Proanthocyanins"], label = "Colour 1")
plt.xlabel("Flavanoids")
plt.ylabel("Proanthocyanins")
plt.legend()
plt.title('Scatter Plot (Original Data)')
plt.grid(True)
plt.show()
```

```python
# standardizing data
scaler = StandardScaler()
df = scaler.fit_transform(data)
```

```python
# applying PCA
pca = PCA(n_components = 2)  # Use 2 components for simplicity, data is 2D
principal_components = pca.fit_transform(df)
```

```python
# Eigenvalues
eigenvalues = pca.explained_variance_
print("Eigenvalues:")
print(eigenvalues)

# Eigenvectors
eigenvectors = pca.components_
print("\nEigenvectors:")
print(eigenvectors)
```

```python
# reduced data
pca_df = pd.DataFrame(data = principal_components, columns = ['Principal Component 1', 'Principal Component 2'])
pca_df["Colour"] = Y 
pca_df
```

```python
# plotting reduced data

import seaborn as sns

# Plot the PCA components
plt.figure(figsize=(10, 6))
sns.scatterplot(x = pca_df['Principal Component 1'], y = pca_df['Principal Component 2'], hue = pca_df["Colour"], s=100)
# Add plot details
plt.title('Scatter Plot (PCA of Wine Dataset)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Colour')
plt.grid(True)
plt.show()
```
